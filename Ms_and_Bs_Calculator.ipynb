```python
# Function to calculate the percentage difference

def porcentaje_diferencia(valor_real, valor_calculado):
    return abs((valor_real - valor_calculado) / valor_real) * 100

# Add the chemical composition of the steel alloy (wt%)

C = 0.804
Si = 1.152
Mn = 0.641
Ni = 0.053
Mo = 0.556
Cr = 1.457
V = 0.231
Co = 0.013
Cu = 0.048
Al = 0.128
W = 0.0067
N = 0
Nb = 0.0058
Ti = 0.011
Ru = 0
B = 0.0024
Fe = 94.8911

##############################################

# EQUATIONS

# Saha Ms = 539-(423*C)-(30,4*Mn)-(17,7*Ni)-(12,1*Cr)-(7,5*Si*Mo*W)
Saha = 539 - (423 * C) - (30.4 * Mn) - (17.7 * Ni) - (12.1 * Cr) - (7.5 * Si) - (7.5 * Mo) - (7.5 * W)

# Andrews Corrected Ms = 512-(453*C)-(16,9*Ni)+(15*Cr)-(9,5*Mo)+(217*(C^2))-(71,5*C*Mn)-(67,6*C*Cr)
Andrews_Corrected = (512 - (453 * C) - (16.9 * Ni) + (15 * Cr) - (9.5 * Mo) + (217 * (C**2)) - (71.5 * C * Mn) - (67.6 * C * Cr))

# Nehenberg Ms = 500-(300*C)-(33*Mn)-(17*Ni)-(22*Cr)-(11*Si)
Nehenberg = 500 - (300 * C) - (33 * Mn) - (17 * Ni) - (22 * Cr) - (11 * Si)

# Steven-Haynes Ms = 561-(474*C)-(33*Mn)-(17*Ni)-(17*Cr)-(21*Si)
Steven_Haynes = 561 - (474 * C) - (33 * Mn) - (17 * Ni) - (17 * Cr) - (21 * Si)

# Payson & Savage Ms = ((930-570*C-60*Mn-50*Cr-30*Ni-20*Si-20*Mo-20*W)-32)/1,8
Payson_Savage = ((930 - 570 * C - 60 * Mn - 50 * Cr - 30 * Ni - 20 * Si - 20 * Mo - 20 * W) - 32) / 1.8

# Rowland Ms = ((930-600*C-60*Mn-50*Cr-30*Ni-20*Si-20*Mo-20*W)-32)/1,8
Rowland = ((930 - 600 * C - 60 * Mn - 50 * Cr - 30 * Ni - 20 * Si - 20 * Mo - 20 * W) - 32) / 1.8

# Grange & Stewart Ms = ((1000-650*C-70*Mn-70*Cr-35*Ni-50*Mo)-32)/1,8
Grange_Stewart = ((1000 - 650 * C - 70 * Mn - 70 * Cr - 35 * Ni - 50 * Mo) - 32) / 1.8

# Andrews Empirical Ms = 539-423*C-30,4*Mn-12,1*Cr-17,7*Ni-7,5*Mo
Andrews_Empirical = 539 - 423 * C - 30.4 * Mn - 12.1 * Cr - 17.7 * Ni - 7.5 * Mo

# Steven & Haynes 2 Ms = 561-474*C-33*Ni-17*Cr-17*Ni-21*Mo
Steven_Haynes2 = 561 - 474 * C - 33 * Ni - 17 * Cr - 17 * Ni - 21 * Mo

############################################################################

# PLOT DIAGRAM

Ms = [
    Saha,
    Andrews_Corrected,
    Nehenberg,
    Steven_Haynes,
    Payson_Savage,
    Rowland,
    Grange_Stewart,
    Andrews_Empirical,
    Steven_Haynes2,
]
equations = [
    "Saha",
    "Andrews Corrected",
    "Nehenberg",
    "Steven-Haynes",
    "Payson & Savage",
    "Rowland",
    "Grange & Stewart",
    "Andrews Empirical",
    "Steven & Haynes 2",
]

# Ask if the user knows the value of Ms
conoce_Ms = input("¿Do you know the experimental value of Ms? (yes/no): ").strip().lower()

plt.figure(figsize=(10, 5))

if conoce_Ms == "yes":
    Ms_real = float(input("Enter the experimental value of Ms: "))
    
    diferencias = [porcentaje_diferencia(Ms_real, ms) for ms in Ms]

    # Percentage difference chart
    bars = plt.bar(equations, diferencias, color="salmon")
    
    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, yval + 0.5, f"{yval:.1f}%", ha='center', va='bottom', fontsize=10, color='black')

    plt.title("Percentage difference between experimental Ms and equations")
    plt.ylabel("Difference (%)")
else:
    # Ms Value Chart
    bars = plt.bar(equations, Ms, color="steelblue")
    
    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, yval + 5, f"{yval:.1f}", ha='center', va='bottom', fontsize=10, color='black')

    plt.title("Ms Calculated by Different Equations")
    plt.ylabel("Ms (°C)")

plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
# Function to calculate the percentage difference

def porcentaje_diferencia(valor_real, valor_calculado):
    return abs((valor_real - valor_calculado) / valor_real) * 100

# Add the chemical composition of the steel alloy (wt%)

C = 0.804
Si = 1.152
Mn = 0.641
Ni = 0.053
Mo = 0.556
Cr = 1.457
V = 0.231
Co = 0.013
Cu = 0.048
Al = 0.128
W = 0.0067
N = 0
Nb = 0.0058
Ti = 0.011
Ru = 0
B = 0.0024
Fe = 94.8911

##############################################

# EQUATIONS

# Bohemen Bs = 839-86*Mn-23*Si-67*Cr-75*Mo-33*Ni-270*(1-EXP(-1,33*C))
Bohemen = 839-86*Mn-23*Si-67*Cr-75*Mo-33*Ni-270*(1-mt.exp(-1.33*C))

# Steven and Hyanes Bs = 830-270*C-90*Mn-37*Ni-70*Cr-83*Mo
Steven_Hyanes = 830-270*C-90*Mn-37*Ni-70*Cr-83*Mo

# Lee & Matt Bs = 745-(110*C)-(59*Mn)-(39*Ni)-(68*Cr)-(106*Mo)+(17*Ni*'Mn)+(6*(Cr^2))+(29*(Mo^2))
Lee_Matt = 745-(110*C)-(59*Mn)-(39*Ni)-(68*Cr)-(106*Mo)+(17*Ni*Mn)+(6*(Cr**2))+(29*(Mo**2))

# Zhao 1 Bs = 630-(45*Mn)-(40*V)-(35*Si)-(30*Cr)-(25*Mo)-(20*Ni)-(15*W)
Zhao1 = 630-(45*Mn)-(40*V)-(35*Si)-(30*Cr)-(25*Mo)-(20*Ni)-(15*W)

# Zhao 2 Bs = 720-585,63*C+126,6*(C^2)-91,68*Mn+7,82*(Mn^2)-0,3378*(Mn^3)-66,34*Ni+6,06*(Ni^2)-0,232*(Ni^3)-31,66*Cr+2,17*(Cr^2)-42,37*Mo+9,16*Co-0,1255*(Co^2)+0,000284*(Co^3)-36,02*Cu-46,15*Ru
Zhao2 = 720-585.63*C+126.6*(C**2)-91.68*Mn+7.82*(Mn**2)-0.3378*(Mn**3)-66.34*Ni+6.06*(Ni**2)-0.232*(Ni**3)-31.66*Cr+2.17*(Cr**2)-42.37*Mo+9.16*Co-0.1255*(Co**2)+0.000284*(Co**3)-36.02*Cu-46.15*Ru

# Bodnaretal Bs = 844-597*C-63*Mn-16*Ni-78*Cr
Bodnaretal = 844-597*C-63*Mn-16*Ni-78*Cr

# Kirkaldy & Venugopalan Bs = 656-57,7*C-75*Si-35*Mn-15,3*Ni-34*Cr-41,2*Mo
Kirkaldy_Venugopalan = 656-57.7*C-75*Si-35*Mn-15.3*Ni-34*Cr-41.2*Mo

# Suehiro Bs = 718-425*C-42,5*Mn
Suehiro = 718-425*C-42.5*Mn

# Lee Bs = 732-202*C+216*Si-85*Mn-37*Ni-47*Cr-39*Mo
Lee = 732-202*C+216*Si-85*Mn-37*Ni-47*Cr-39*Mo

# Kunitake & Okada Bs = 711-361,9*C+261,9*(C^2)-28,3*Mn+43,7*Si
Kunitake_Okada = 711-361.9*C+261.9*(C**2)-28.3*Mn+43.7*Si

# J Trzaska Bs = 771-231,5*C-69*Mn-58,5*Cr-31*Ni-23*Si-55*Mo-41*V
J_Trzaska = 771-231.5*C-69*Mn-58.5*Cr-31*Ni-23*Si-55*Mo-41*V

############################################################################

# PLOT DIAGRAM

Bs = [
    Bohemen,
    Zhao1,
    Zhao2,
    Bodnaretal,
    Kirkaldy_Venugopalan,
    Suehiro,
    Lee,
    Kunitake_Okada,
    J_Trzaska,
]
equations = [
    "Bohemen",
    "Zhao 1",
    "Zhao 2",
    "Bodnaretal",
    "Kirkaldy and Venugopalan",
    "Suehiro",
    "Lee",
    "Kunitake & Okada",
    "J Trzaska",
]

# Ask if the user knows the value of Bs
conoce_Bs = input("¿Do you know the experimental value of Bs? (yes/no): ").strip().lower()

plt.figure(figsize=(10, 5))

if conoce_Bs == "yes":
    Bs_real = float(input("Enter the experimental value of Bs: "))
    
    diferencias = [porcentaje_diferencia(Bs_real, bs) for bs in Bs]

    # Percentage difference chart
    bars = plt.bar(equations, diferencias, color="salmon")
    
    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, yval + 0.5, f"{yval:.1f}%", ha='center', va='bottom', fontsize=10, color='black')

    plt.title("Percentage difference between experimental Bs and equations")
    plt.ylabel("Difference (%)")
else:
    # Bs Value Chart
    bars = plt.bar(equations, Bs, color="steelblue")
    
    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, yval + 5, f"{yval:.1f}", ha='center', va='bottom', fontsize=10, color='black')

    plt.title("Bs Calculated by Different Equations")
    plt.ylabel("Bs (°C)")

plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```
